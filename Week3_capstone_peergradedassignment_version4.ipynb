```python
# what's the title
soup.title
```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table = soup.find('table', class_='wikitable sortable')
right_table
```

#### Find the rows there are 3 columns, so we set up three empty lists A B C and Import Pandas 

```python

import pandas as pd

A = []
B = []
C = []
for row in right_table.findAll('tr'):
    cells = row.findAll('td')
    if len(cells) == 3: 
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
df=pd.DataFrame(A,columns = ['Postal Code'])
df['Borough'] = B
df['Neighborhood'] = C 

df
```

#### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
# If you want to see something cool: neWdf = df[df.Borough == 'Not assigned\n'].reset_index(drop=True)
# then -> neWdf
```

```python
neWdf = df[df.Borough != 'Not assigned\n'].reset_index(drop=True)
neWdf

#neWdf = df[df.Borough != 'Not assigned\n'].reset_index(inplace=True)
#neWdf
```

```python
neWdf.shape
```

#### A little more clean up , stuff like \n 

```python
neWdf = neWdf.replace({'\n': '<br>'}, regex=True)
```

```python
neWdf
```

```python
neWdf = neWdf.replace({'<br>': ''}, regex=True)
```

```python
neWdf
```

#### More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma

```python
neWdf = neWdf.replace({'/': ','}, regex=True)
neWdf
```

#### Group all with the same postal code

```python
neWdf3 = neWdf.groupby(["Postal Code", "Borough"])["Neighborhood"].apply(", ".join).reset_index()
neWdf3.head()
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

address = "Toronto, ON"

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto city are {}, {}.'.format(latitude, longitude))

```

```python
# create map of Toronto using latitude and longitude values
import folium 
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
map_toronto
```

#### Plot neighborhood and borough to toronto map 

```python
for lat, lng, borough, neighborhood in zip(
        neWdf5['Latitude'], 
        neWdf5['Longitude'], 
        neWdf5['Borough'], 
        neWdf5['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  

map_toronto
```
